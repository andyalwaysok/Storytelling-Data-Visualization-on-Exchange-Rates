The dataset we'll use describes Euro daily exchange rates between 1999 and 2021.

If the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro

Goal of this project will be on the exchange rate between the euro and the American dollar.

And also to create a storytelling data visualization using the data we have

```python
import pandas as pd

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv', encoding='latin-1')
```

```python
exchange_rates.columns
```

```python
exchange_rates.rename({'Period\\Unit:': 'time', '[US dollar ]':'us_dollar'}, 
                      axis=1, inplace=True)
```

```python
exchange_rates.columns
```

```python
exchange_rates['time'] = pd.to_datetime(exchange_rates['time'])
```

```python
exchange_rates.info()
```

```python
exchange_rates.head()
```

```python
exchange_rates.sort_values('time', inplace=True)
```

```python
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['time', 'us_dollar']]
```

```python
euro_to_dollar.head()
```

```python
euro_to_dollar.us_dollar.value_counts()
```

After investigating US Dollar column, we noticed that 62 values are under "-". 

We want to drop them.

```python
mask = euro_to_dollar.us_dollar == '-'
euro_to_dollar = euro_to_dollar[~mask]
euro_to_dollar.us_dollar.value_counts()
```

```python
euro_to_dollar.us_dollar.astype(float)
```

```python
euro_to_dollar
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar.rolling(30).mean()
euro_to_dollar
```

```python
euro_to_dollar.head(30)
```

We will show how the euro-dollar rate has changed during the coronavirus pandemic. We can show the 2020 data and the 2016-2019 data as a baseline.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
euro_to_dollar['year'] = euro_to_dollar.time.dt.year
```

```python
euro_to_dollar.head()
```

```python
mask_2020 = euro_to_dollar.year >= 2020
euro_to_dollar_2020 = euro_to_dollar[mask_2020]
euro_to_dollar_2020
```

```python
euro_to_dollar_rest = euro_to_dollar[(euro_to_dollar['year'] >= 2016) & (euro_to_dollar['year'] <= 2019)]
euro_to_dollar_rest
```

```python
plt.plot(euro_to_dollar_rest.time, euro_to_dollar_rest.rolling_mean)
plt.title('Dollar to Euro From 2016-2019')
plt.xlabel('Time')
plt.ylabel('Dollar')
plt.xticks(rotation=45)
plt.show()

plt.plot(euro_to_dollar_2020.time, euro_to_dollar_2020.rolling_mean)
plt.title('Dollar to Euro From 2020')
plt.xlabel('Time')
plt.ylabel('Dollar')
plt.xticks(rotation=45)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14,8))

ax1.plot(euro_to_dollar_rest.time, euro_to_dollar_rest.rolling_mean, color='Red', alpha=0.5, linewidth=3)

for location in ['left', 'right', 'top', 'bottom']:
    ax1.spines[location].set_visible(False)

ax1.tick_params(bottom=False, left=False) 
ax1.tick_params(axis='x', rotation=45) 

ax1.text(x=735964, y=1.25, s='Dollar to Euro between 2016-2019', weight='bold', size=17) 

ax2.plot(euro_to_dollar_2020.time, euro_to_dollar_2020.rolling_mean, color='Blue', alpha=0.5, linewidth=3)

for location in ['left', 'right', 'top', 'bottom']:
    ax2.spines[location].set_visible(False)

ax2.tick_params(bottom=False, left=False) 
ax2.tick_params(axis='x', rotation=45) 


ax2.text(x=737425, y=1.227, s='Dollar to Euro in 2020', weight='bold', size=17) 


plt.show()
```
